```python
df = pd.read_csv('../../week_1/mysolutions/co2_mm_mlo.txt',comment='#',header=None, delim_whitespace=True)
```

```python
mydf = df.loc[:,[0,1,3]] # call all
mydf.columns = ['year','month','CO2']
mydf.head()
```

```python
mydf['CO2'] = np.where(mydf['CO2']==-99.99, np.NaN,mydf['CO2'])
```

```python
from matplotlib.cm import get_cmap

cmap = get_cmap('inferno')
```

```python
fig = plt.figure(figsize=(12,8))
# line_norm = Normalize(1, 13)
for mnth in range(1,13):
    plt.plot(mydf.year[mydf.month==mnth], mydf.CO2[mydf.month==mnth], 'o', 
             markersize=12, color = cmap(mnth/13))
```

```python
fig = plt.figure(figsize=(12,8))
```

```python
mydf_avg  = mydf.groupby('year').mean()
mydf_avg.head()
```

```python
mydf_avg.CO2.plot(kind='line')
```

```python
fig = plt.figure(figsize=(12,8))
# line_norm = Normalize(1, 13)
for mnth in range(1,13):
    plt.plot(mydf.year[mydf.month==mnth], mydf.CO2[mydf.month==mnth], 'o', 
             markersize=12, color = cmap(mnth/13));
mydf_avg.CO2.plot(kind='line',color = 'green');

plt.title('CO2 index by month (dots) & yearly avg (blue line)')
plt.xlabel('Year')
plt.ylabel('CO2 Values')

plt.ylim(300,420)
plt.xlim(1975,2020)

plt.savefig('week2_warm1.jpg')
```
