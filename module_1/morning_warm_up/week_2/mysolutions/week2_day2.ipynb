```python
import pandas as pd
```

```python
df = pd.read_csv('master.csv')
```

```python
df.head()
```

```python
df[df['country'].apply(lambda x: x.startswith('U'))]
```

```python
df[~df['country'].apply(lambda x: x.startswith('U'))]['country'].unique()
```

```python
df_us = df[df['country']=='United States']
```

```python
df_us['age'] = df_us['age'].astype('category')
```

```python
cat = df_us['age'].unique()
print(cat)
newcat = cat.reorder_categories(['5-14 years', '15-24 years','25-34 years',
                        '35-54 years' , '55-74 years','75+ years' ])
```

```python
df_us['age'] = df_us['age'].astype(newcat)
```

```python
df_us.sort_values('age')
```

# Draw a barplot of # suicides(y) per age-group (x) for 2015

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,8))
ax = sns.barplot(x='age',y='suicides_no', data=df_us)
ax.set_title('Suicide in the US by age group')
ax.set_ylabel('Suicides')
ax.set_xlabel('Age Group')
ax.grid(True)
plt.savefig('visualizations/suicides_barplot.png')
```
